####movielens/100k-ratings
This dataset contains 100,000 ratings from 943 users on 1,682 movies. This dataset is the oldest version of the MovieLens dataset.

```python
# Getting ratings dataset.
ratings = tfds.load("movielens/100k-ratings", split="train")
```

```python
# Manual data lookup of rattings data.
# It is implicit feedback from the user, watched movies as positivie and unwatchedd a negative.
# Explicit feedback is for movie rattings, provided by the user saying how much user likes watching the movie.
for x in ratings.take(2).as_numpy_iterator():
  pprint.pprint(x)
```

```python
# Creating mapping of user id, movie, and rating.
ratings = ratings.map(lambda x: {
    "movie_title": x["movie_title"],
    "user_id": x["user_id"],
    "user_rating": x["user_rating"]
})

for x in ratings.take(10).as_numpy_iterator():
  pprint.pprint(x)
```
